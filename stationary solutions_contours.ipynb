### Stationary solutions for spherically symmetric accretion onto a star

$$\mathcal{M}^2=\frac{u^2}{c_s^2}\\\xi=\frac{r}{r_0}$$
$$\frac{\mathcal{M}^2-1}{2\mathcal{M}^2}\frac{d\mathcal{M}^2}{d\xi}=[1+(\frac{\gamma-1}{2})\mathcal{M}^2]\cdot[\frac{2E}{\xi}-(\frac{5-3\gamma}{\gamma-1})\frac{GM}{2r_0\xi^2}]/[E+\frac{GM}{r_0\xi}]=F(\mathcal{M}^2,\xi)$$

$$\delta f(\mathcal{M}^2)=g(\xi)$$
$$f(\mathcal{M}^2)=(\frac{1}{\gamma-1}+\frac{\mathcal{M}^2}{2})/\mathcal{M}^{2\frac{\gamma-1}{\gamma+1}}$$
$$g(\xi)=(\frac{1}{\gamma-1}-\frac{3}{2}+\frac{2}{\xi})\cdot\xi^{4\frac{\gamma-1}{\gamma+1}}$$
$$\delta=(\frac{\beta_0}{\beta})^{2\frac{\gamma-1}{\gamma+1}}=e^{\frac{2}{\gamma+1}\cdot\frac{S-S_0}{C_V}}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# x: xi; y: mach_number^2; gamma = 4/3; delta = 1
def F(x, y, gamma):
    fy = ((1/(gamma-1))+y/2)/y**((gamma-1)/(gamma+1))
    gx = ((1/(gamma-1))-(3/2)+(2/x))*x**(4*(gamma-1)/(gamma+1))
    return gx/fy

gamma = 4/3
#drawing contour of the implicit function
x = np.linspace(0.1, 3, 100)                                                    
y = np.linspace(0.1, 4, 100)
X, Y = np.meshgrid(x, y)
Z = F(X, Y, gamma=4/3)**((gamma-1)/(gamma+1))
Z1 = Z.copy()
Z1[Z1>1.04] = None

plt.subplots(figsize=(12,10), dpi = 72)
plt.contourf(X, Y, Z, 16, colors = "white")
contours = plt.contour(X, Y, Z1, 12, colors = 'black')
#Mark the contour lines corresponding to different values of delta.
#When delta = 1, we have the Bondi accretion and Parker wind.
plt.clabel(contours, inline=True, fontsize=14*1.5, manual= [(1.5, 1.0), (2.0, 1.0), (2.5, 1.0), (0.6, 1.0), (0.75, 1.0), (0.25, 1.0), (0.4, 1.0), (1,0.5),(1.0, 0.29), [1.0, 2.0], [1.0, 2.75], [1.0, 3.5], [1.0, 3.75]])
contours = plt.contour(X, Y, Z1, 1, colors='black', linewidths= 3)
plt.clabel(contours, inline=True, fontsize=14*1.5, manual= [(1.6,1.6),(1.25,0.7)])
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.xlabel(r'$\xi=r/r_s$', fontsize=12*2)
plt.ylabel(r'$\mathcal{M}^2 = v^2 / c_s^2$', fontsize=12*2)

plt.text(0.78, 1.50, r'$\mathbf{1}$', fontsize=15*2)
plt.text(1.5, 1.50, r'$\mathbf{2}$', fontsize=15*2) 
plt.text(0.97, 0.65, r'$\mathbf{3}$', fontsize=15*2)
plt.text(0.97, 2.2, r'$\mathbf{4}$', fontsize=15*2)
plt.text(2.29, 1.45, r'$\mathbf{5}$', fontsize=15*2)
plt.text(0.40, 1.45, r'$\mathbf{6}$', fontsize=15*2)
#Sonic point (1,1)
plt.plot(1, 1, 'ro')
```
